```python
import os
```

**Огонёк**


```python
dirpath = 'огонек_1'
for filename in os.listdir(dirpath):
  filepath = os.path.join(dirpath, filename)
  if os.path.isfile(filepath):
    try:
      with open(filepath, encoding = 'utf-8') as fd:
        text = fd.read()
        text = text.lower()
        text = re.sub(r'@', r' ', text)
        text = re.sub(r'\|', r' ', text)
        text = re.sub(r'-\n', r'', text)
        text = re.sub(r'\n', r' ', text)
        text = re.sub(r'\x0c', r' ', text)
        text = re.sub(r'\u00AD ', r'', text)
        text = re.sub(r'(\w)(—)', r'\1 \2', text)
        text = re.sub(r'(\W)(\w)', r'\1 \2', text)
        text = re.sub(r'(\w)(-)(\s)(\w)', r'\1\2\4', text)
        #text = re.sub(r'-', r'', text)
        print(text[:100])
        with open ('ogonek_pre_add_2.txt', 'a') as file:
          file.write(text)
    except:
       print(f'с файлом {filepath} проблема')         
```

**Спортивная жизнь, Сельская молодежь**
