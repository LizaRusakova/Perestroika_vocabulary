```python
from pymorphy2 import MorphAnalyzer
morph = MorphAnalyzer()
```

```python
noun_data = pd.read_csv('noun_data_pre.csv')
verb_data = pd.read_csv('verb_data_pre.csv')
adj_data = pd.read_csv('adj_data_pre.csv')
```

```python
adj_data
```

```python
noun_data = noun_data.drop(['Unnamed: 0.1', 'Unnamed: 0'], axis = 'columns')
verb_data = verb_data.drop(['Unnamed: 0.1', 'Unnamed: 0'], axis = 'columns')
adj_data = adj_data.drop(['Unnamed: 0.1', 'Unnamed: 0'], axis = 'columns')
```

```python
noun_list = noun_data['0'].to_list()
verb_list = verb_data['0'].to_list()
adj_list = adj_data['0'].to_list()
```

```python
adj_list
```

```python
len(verb_list)
```

```python
noun_list_final = []
verb_list_final = []
adj_list_final = []
```

```python
for word in noun_list:
  if type(word) == str:
    normal_form = morph.parse(word)[0].normal_form
    noun_list_final.append(normal_form)
for word in verb_list:
  if type(word) == str:
    normal_form = morph.parse(word)[0].normal_form
    verb_list_final.append(normal_form)
for word in adj_list:
  if type(word) == str:
    normal_form = morph.parse(word)[0].normal_form
    adj_list_final.append(normal_form)      
```

```python
noun_data_clean = pd.Series(noun_list_final)
verb_data_clean = pd.Series(verb_list_final)
adj_data_clean = pd.Series(adj_list_final)
```

```python
noun_tags = []
verb_tags = []
adj_tags = []
```

```python
for word in noun_list_final:
  tag = morph.parse(word)[0].tag.POS
  noun_tags.append(tag)
for word in verb_list_final:
  tag = morph.parse(word)[0].tag.POS
  verb_tags.append(tag)
for word in adj_list_final:
  tag = morph.parse(word)[0].tag.POS
  adj_tags.append(tag)      
```

```python
noun_data_clean = pd.DataFrame({"Word": noun_list_final,
                   "Tag": noun_tags})
verb_data_clean = pd.DataFrame({"Word": verb_list_final,
                   "Tag": verb_tags})
adj_data_clean = pd.DataFrame({"Word": adj_list_final,
                   "Tag": adj_tags})
```

```python
verb_data_clean
```

```python
all_data = noun_data_clean.append(verb_data_clean, ignore_index = True)
all_data = all_data.append(adj_data_clean, ignore_index = True)
```

```python
noun_data_final = all_data[all_data['Tag'] == 'NOUN']
verb_data_final = all_data[all_data['Tag'] == 'INFN']
adj_data_final = all_data[all_data['Tag'] == 'ADJF']
```

```python
noun_data_final
```

```python
verb_data_final
```

```python
adj_data_final
```

```python
noun_data_final.to_csv('noun_data_pre_final.csv', encoding = 'utf-8', index = False)
verb_data_final.to_csv('verb_data_pre_final.csv', encoding = 'utf-8', index = False)
adj_data_final.to_csv('adj_data_pre_final.csv', encoding = 'utf-8', index = False)
```

**Дальше считаем частотность**

```python
noun = noun_data_final['Word'].to_list()
verb = verb_data_final['Word'].to_list()
adj = adj_data_final['Word'].to_list()
```

```python
from collections import Counter 
```

```python
noun_data = pd.DataFrame({'Слово': dict(Counter(noun)).keys(), 'Кол-во употреблений': dict(Counter(noun)).values()})
noun_data['Часть речи'] = 'NOUN'
adj_data = pd.DataFrame({'Слово': dict(Counter(adj)).keys(), 'Кол-во употреблений': dict(Counter(adj)).values()})
adj_data['Часть речи'] = 'ADJ'
verb_data = pd.DataFrame({'Слово': dict(Counter(verb)).keys(), 'Кол-во употреблений': dict(Counter(verb)).values()})
verb_data['Часть речи'] = 'VERB'
```

```python
noun_data.to_csv('noun_data_pre_counter.csv', encoding = 'utf-8', index = False)
verb_data.to_csv('verb_data_pre_counter.csv', encoding = 'utf-8', index = False)
adj_data.to_csv('adj_data_pre_counter.csv', encoding = 'utf-8', index = False)
```
